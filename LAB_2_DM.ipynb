#Statistical summary of the dataset

```python
import pandas as pd
df = pd.read_csv("/content/titanic.csv")
su = df.describe()
print(su)
```

```python
print(df['Age'].describe())
```

#Create Visualization using Matplotlib


Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'].dropna(), bins=20, edgecolor='orange', color='red')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Bar chart

```python
plt.figure(figsize=(6, 4))
df['Pclass'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

Box plot

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Age'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

#Create Visualization using Seaborn

Pair plot

```python
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df[['Age','Fare','Pclass','Survived']].dropna())
plt.show()
```

Heat map

```python
correlation_matrix = df[['Age', 'Fare', 'SibSp', 'Parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

Count plot

```python
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

Violin plot

```python
sns.violinplot(x='Survived', y='Age', data=df, palette='muted')
plt.title('Age Distribution by Survived')
plt.show()
```
