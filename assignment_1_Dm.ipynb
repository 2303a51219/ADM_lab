```python
import pandas as pd
df = pd.read_csv('/content/bank_transactions_data_2.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['AccountBalance'].fillna(df['AccountBalance'].mean(), inplace=True)
```

```python
df['AccountBalance'].fillna(df['AccountBalance'].median(), inplace=True)
```

```python
df['IP Address'].fillna(df['IP Address'].mode()[0], inplace=True)
```

```python
print(df.isnull().sum())
```

Min-max normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns_to_normalize = ['CustomerAge','AccountBalance','TransactionDuration']
df[columns_to_normalize]=scaler.fit_transform(df[columns_to_normalize])
df[columns_to_normalize].head()
```

Standardization


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[columns_to_normalize]=scaler.fit_transform(df[columns_to_normalize])
df[columns_to_normalize].head()
```
